```python
# Calcule a aptidão máxima com base na população, aptidão máxima de 28
def calculate_max_fitness():
    i = board_size - 1
    max_fitness = 0
    while i > 0:
        max_fitness += i
        i -= 1
    return max_fitness

# Gera um indivíduo aleatório com base no tamanho do tabuleiro
def random_individual(board_size):
    return [ random.randint(0,board_size-1) for _ in range(board_size) ]

# Define a aptidão de um indivíduo com base no número de rainhas que não atacam checando a diagonal esquerda e direita e a horizontal.
def fitness(individual):
    horizontal_collisions = sum([individual.count(queen)-1 for queen in individual])/2
    diagonal_collisions = 0

    n = len(individual)
    left_diagonal = [0] * 2*n
    right_diagonal = [0] * 2*n
    for i in range(n):
        left_diagonal[i + individual[i] - 1] += 1
        right_diagonal[len(individual) - i + individual[i] - 2] += 1

    for i in range(2*n-1):
        counter = 0
        if left_diagonal[i] > 1:
            counter += left_diagonal[i]-1
        if right_diagonal[i] > 1:
            counter += right_diagonal[i]-1
        diagonal_collisions += counter / (n-abs(i-n+1))
    return int(max_fitness - (horizontal_collisions + diagonal_collisions))

# Calcule a probabilidade de um indivíduo com base em sua aptidão em comparação com max_fitness que é maximizado e o máx é 28.
def probability(individual, fitness):
    return fitness(individual) / max_fitness

# Pegue um indivíduo aleatório com base na população por probabilidade
def random_pick(population, probabilities):
    populationpro = zip(population, probabilities)
    total = sum(w for c, w in populationpro)
    r = random.uniform(0, total)
    upto = 0
    for c, w in zip(population, probabilities):
        if upto + w >= r:
            return c
        upto += w
    assert False 

# Retorne um novo indivíduo, seus genes são de 2 pais x e y com um aleatório 
def crossover(x, y):
    n = len(x)
    #c = 4
    c = random.randint(0, n - 1)
    return x[0:c] + y[c:n]

# Retorna um indivíduo com um gene aleatório modificado entre 1 e n do tamanho do tabuleiro
def mutate(x):
    n = len(x)
    c = random.randint(0, n - 1)
    m = random.randint(1, n)
    x[c] = m
    return x

#Criação de novas populações com indivíduos aleatórios 
def genetic_queen(population, fitness):
    #Probabilidade de mutação
    new_population = []
    probabilities = [probability(n, fitness) for n in population]
    for i in range(len(population)):
        x = random_pick(population, probabilities)
        y = random_pick(population, probabilities)
        child = crossover(x, y)
        if random.random() < mutation_probability:
            child = mutate(child)
        new_population.append(child)
        if fitness(child) == max_fitness: break
    return new_population
```
