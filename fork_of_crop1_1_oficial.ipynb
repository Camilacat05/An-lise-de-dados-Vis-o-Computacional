```python
def recorte (xc,yc,h,w,image, mask):
    rec1 = []
    rec2 = []

    for i in range(0,20):
        x =  random.randint(-30,30)
        y = random.randint(-30,30)
        
        new_cx = xc - x                                     
        new_cy = yc - y
        crop = image[new_cx-int(h/2):new_cx+int(h/2), new_cy-int(w/2):new_cy+int(w/2)]
        cropm = mask[new_cx-int(h/2):new_cx+int(h/2), new_cy-int(w/2):new_cy+int(w/2)]
        
        filename = 'IMG-0001-00046'
        name1 = '%s_corte1_%s.png' % (filename, str(i))
        imsave('./'+name1, crop)
        filename1 = 'Mask-0001-00046'
        name2 = '%s_corte1_%s.png' % (filename1, str(i))
        imsave('./'+name2, cropm)
      #  name2 = '%s_corte2_%s.png' % (filename, str(i))3
      #  imsave('./'+name2, crop1)
      #  name3 = '%s_corte3_%s.png' % (filename, str(i))30
      #  imsave('./'+name3, crop2)
     #   name4 = '%s_corte4_%s.png' % (filename, str(i))
      #  imsave('./'+name4, crop3)
        
        rec1.append(crop)
        rec2.append(cropm)
      #  rec2.append(crop1)
      #  rec3.append(crop2)
      #  rec4.append(crop3)
        
    return rec1, rec2
```

```python
recorte1,recorte2 = [],[]
xc, yc = 278,170
h, w = 64, 64 # 32, 64

for i in range(1):
    rec1,rec2 = recorte(xc,yc,h,w,image,mask)
    
    recorte1.append(rec1)
    recorte2.append(rec2)
  #  recorte3.append(rec3)
   # recorte4.append(rec4)
```
